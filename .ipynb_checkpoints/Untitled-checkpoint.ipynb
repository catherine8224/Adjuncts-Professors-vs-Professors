```python
#Adjunct Profs Data
df_2016a = pd.read_csv('editors2016A.csv', header=None, names = ['Name', 'Agency', 'Total Pay', 'School','School1', 'Title', 'Rate of Pay', 
            'Pay Year', 'Pay Basis', 'Branch/Major Category'])
df_2017a = pd.read_csv('editors2017A.csv')
df_2018a = pd.read_csv('editors2018A.csv', header=None, skiprows=1, names =['Name', 'Agency', 'Total Pay', 'School','School1', 'Title', 'Rate of Pay', 
            'Pay Year', 'Pay Basis', 'Branch/Major Category'])

#get rid of $ and , and set the values in the column as integers
df_2016a['Total Pay'] = df_2016a['Total Pay'].replace({'\$':''}, regex = True).replace({',':''}, regex = True)
df_2016a['Total Pay'] = df_2016a['Total Pay'].astype(int)
#df_2016a

df_2016a['Rate of Pay'] = df_2016a['Rate of Pay'].replace({'\$':''}, regex = True).replace({',':''}, regex = True)
df_2016a['Rate of Pay'] = df_2016a['Rate of Pay'].astype(int)

df_2017a['Total Pay'] = df_2017a['Total Pay'].replace({'\$':''}, regex = True).replace({',':''}, regex = True)
df_2017a['Total Pay'] = df_2017a['Total Pay'].astype(int)

df_2017a['Rate of Pay'] = df_2017a['Rate of Pay'].replace({'\$':''}, regex = True).replace({',':''}, regex = True)
df_2017a['Rate of Pay'] = df_2017a['Rate of Pay'].astype(int)

df_2018a['Total Pay'] = df_2018a['Total Pay'].replace({'\$':''}, regex = True).replace({',':''}, regex = True)
df_2018a['Total Pay'] = df_2018a['Total Pay'].astype(int)

df_2018a['Rate of Pay'] = df_2018a['Rate of Pay'].replace({'\$':''}, regex = True).replace({',':''}, regex = True)
df_2018a['Rate of Pay'] = df_2018a['Rate of Pay'].astype(int)

#limit data based on whether total pay > rate of pay, and whether pay is annual
df_2016a = df_2016a[df_2016a['Total Pay']>=df_2016a['Rate of Pay']]
df_2016a = df_2016a[df_2016a['Pay Basis']=='Annual']
df_2016a

df_2017a = df_2017a[df_2017a['Total Pay']>=df_2017a['Rate of Pay']]
df_2017a = df_2017a[df_2017a['Pay Basis']=='Annual']

df_2018a = df_2018a[df_2018a['Total Pay']>=df_2018a['Rate of Pay']]
df_2018a = df_2018a[df_2018a['Pay Basis']=='Annual']

#sum total salary based on year
Total2016a = df_2016a['Total Pay'].sum()
Total2017a = df_2017a['Total Pay'].sum()
Total2018a = df_2018a['Total Pay'].sum()

Totals=[]
Totals.extend([Total2016a, Total2017a, Total2018a])
Totals

#Totals.append(Total2016a)
#Totals.append(Total2017a)
#Totals.append(Total2018a)
```
